## 1. Numpy

**What is NumPy?**

NumPy = Numerical Python

It provides:

- Fast numerical arrays
- Vectorized math operations
- Linear algebra & statistics
- At the core is the ndarray (n-dimensional array).

**Why NumPy is Important for Data Analysis**

 - Faster than Python lists
 - Foundation for Pandas, SciPy, Scikit-learn
 - Handles large datasets efficiently

````python
import numpy as np

#Creating Arrays
#From Python lists
a = np.array([1, 2, 3, 4])
b = np.array([[1, 2], [3, 4]])

print(a,b, sep='\n')
````

````python
#Special arrays
zeroes = np.zeros(5)
ones = np.ones((2, 3))
rang = np.arange(0, 10, 2)
lin_space = np.linspace(0, 1, 5)

print(zeroes, ones, rang, lin_space, sep='\n')
````

````python
#Array Properties
a = np.array([[1, 2], [3, 4]])
shape = a.shape     # dimensions
dim = a.ndim      # number of dimensions
size = a.size      # total elements
dtype = a.dtype     # data type

print(f'shape:{shape}\ndimension:{dim}\nsize:{size}\nData_type:{dtype}')
````

***Indexing & Slicing***

````python
a = np.array([1, 2, 3, 4])
b = np.array([[1, 2], [3, 4]])

#1D
print(a[0])
print(a[1:3])

#2D
print(b[0, 1])
print(b[:, 0] )   # all rows, first column
print(b[1, :])    # second row
````

#### Vectorized Operations (Very Important)

No loops needed 

````python
a = np.array([1, 2, 3, 4])

print(f'''{a + 10}
{a * 2}
{a ** 2}
{np.sqrt(a)}''')
````

````python
#Comparison:
a = np.array([1, 2, 3, 4])
a > 2
````

````python
#Aggregation & Statistics

#1D
a = np.array([1, 2, 3, 4])
print(f'''For 1D Array\nAddtion:{a.sum()}
Average:{a.mean()}
Mini-value:{a.min()}
Max_value:{a.max()}
Standard_Deviaton:{a.std()}''')

#For 2D arrays:
b = np.array([[1, 2], [3, 4]])

print(f'''\nFor 2D Array:\nAddtion for col:{b.sum(axis=0)}
Addtion  for row:{b.sum(axis=1)}
Average for col:{b.mean(axis=0)}
Average for row:{b.mean(axis=1)}
Mini-value for col:{b.min(axis=0)}
Mini-value for row:{b.min(axis=1)}
Max_value for col:{b.max(axis=0)}
Max_value for row:{b.max(axis=1)}
Standard_Deviaton for col:{b.std(axis=0)}
Standard_Deviaton for row:{b.std(axis=1)}''')
````

````python
#Boolean Indexing (Filtering Data)
data = np.array([10, 15, 20, 25, 30])

print('Age above 20:',data[data > 20])

#Multiple conditions:

print('Age between 15-30:', data[(data > 15) & (data < 30)])
````

***Handling Missing Data***
 - NumPy uses NaN


````python
data = np.array([1, 2, np.nan, 4])

print(np.isnan(data)) #checking for null values
print(np.nanmean(data) )#ignores null values and provides average of array
print(np.nan_to_num(data))   # replaces NaN with 0 by default
````

***Reshaping Arrays***

````python
a = np.arange(12)

print(a)
print(a.reshape(3, 4))
print(a.reshape(-1, 3))   # auto size

#Flatten:

print(a.ravel())
````

***Broadcasting***

Allows operations between different shapes.


````python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

a + np.array([10, 20, 30])
````

````python
#Random Numbers (Data Simulation)
print(np.random.rand(5))
print(np.random.randn(5))
print(np.random.randint(0, 10, size=5))
````

***Set seed (reproducibility):***

This sets the random seed for NumPy’s random number generator.

What it does:

Makes all NumPy random operations reproducible

The same “random” numbers will be generated every time you run the code

````python
#Set seed (reproducibility)

print(np.random.rand(5))
np.random.seed(42)
print(np.random.rand(5))
````

````python
#Linear Algebra Basics
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(np.dot(A, B)) #Matrix multiplication
print(np.linalg.inv(A)) #Matrix inverse
print(np.linalg.det(A)) #Determinant of matrix
````

## 2. Pandas

**What is Pandas?**

 - Pandas is a Python library used to:
 - Load data (CSV, Excel, SQL, JSON)
-- Clean messy data
 - Analyze & summarize datasets
 - Prepare data for visualization & ML

````python
import pandas as pd

#Series (1D)
s = pd.Series([10, 20, 30], name="scores")
print(s)


# Creating new DataFrame (2D table)
df = pd.DataFrame({
    "name": ["A", "B", "C"],
    "age": [18, 19, 17],
    "score": [85, 90, 88]
})
print()
print(df)
````

````python
#saving Dataframe
df.to_csv(r'data.csv',index=False) #csv

#Loading Data
df = pd.read_csv("data.csv")
df
````

````python
# preview of Data frame
print(df.head(1)) # returns first 1 rows
print()
print(df.tail(1) )# return last 1 rows
print()
print(df.info()) # returns Info about df
print()
print(df.shape) # shape of df
````

***Selecting Data***

````python
#column Wise
print(df["score"])
print(df[["name", "score"]])
````

````python
#Row_wise
print(df.loc[0])         # label-based
print()
print(df.iloc[0] )       # index-based
print()
print(df.loc[0:2, ["name", "age"]])
````

***Filtering Data***

````python
print(df[df["score"] > 85])

print('multiple conditions:')
print(df[(df["age"] >= 18) & (df["score"] > 85)])
````

***Adding & Modifying Columns***

````python
df["passed"] = df["score"] >= 80
df["bonus"] = df["score"] * 0.05
print(df)
````

***Sorting Data***

````python
df.sort_values("age", ascending=True)
df.reset_index()
````

***Handling Missing Data***

````python
#checking
print(df.isna())
print(df.isna().sum())
````

````python
#Drop NUll values
df.dropna() # drops columns
df.dropna(axis=1) #drops rows
````

````python
# Imputing Null Values
df.fillna(0) #fill with Zeros
df["score"].fillna(df["score"].mean()) # fill with average of column
````

***GroupBy (Very Important)***

Used for aggregation.

````python
print(df.groupby("age")["score"].mean())

print()
#Multiple aggregations:

df.groupby("age").agg(
    avg_score=("score", "mean"),
    max_score=("score", "max")
)
````

***Descriptive Statistics***

````python

print(f'''{df.describe()}\n\n
{df["score"].mean()}\n\n
{df["score"].value_counts()}\n\n
{df["age"].unique()}''')
````

***Apply & Map***

````python
df["grade"] = df["score"].apply(
    lambda x: "A" if x >= 90 else "B"
)
df
````

## 3. Visualization
`Matplotlib` – Core Plotting Library

Matplotlib is the foundation for almost all Python plotting. It’s very flexible but can be a bit verbose.

`Seaborn` – Beautiful Statistical Plots

Seaborn is built on Matplotlib and makes plots prettier and easier, especially for datasets.

***Matplotlib vs Seaborn***

| Feature           | Matplotlib          | Seaborn                   |
| ----------------- | ------------------- | ------------------------- |
| Flexibility       | Very high           | Medium                    |
| Default style     | Basic               | Pretty, ready for reports |
| Statistical plots | Needs manual coding | Built-in                  |
| Learning curve    | Steeper             | Easier                    |


````python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

#Basic line plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)
plt.title("Line Plot Example")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

#customized plot
plt.plot(x, y, color='purple', linestyle='--', marker='o', linewidth=2, markersize=8)
plt.grid(True)
plt.title("Customized Line Plot")
plt.show()
````

````python
#Load Sample Dataset
df = sns.load_dataset("tips")  # Example dataset
df.head()
````

***Histogram / Distribution Plot***

````python
sns.histplot(df["total_bill"], bins=20, kde=True)
plt.title("Total Bill Distribution")
plt.show()
````

***Box Plot***

````python
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Total Bill by Day")
plt.show()
````

***Scatter Plot***

````python
sns.scatterplot(x="total_bill", y="tip", data=df, hue="sex", style="time", size="size")
plt.title("Total Bill vs Tip")
plt.show()
````

***Bar Plot***

````python
sns.barplot(x="day", y="total_bill", data=df, ci=None)
plt.title("Average Total Bill per Day")
plt.show()
````
